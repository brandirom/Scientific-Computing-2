```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
alpha = 1e-6
beta = 5.6e-5
T_inf = 50.0
Time_final = 1000.0 
N = 100
hx = 1.0/N
ht = hx*hx/8/alpha
nStepsInTime = int(Time_final / ht)
T = np.zeros((nStepsInTime+1,N+2))
x = np.linspace(0,1,N+2)
for n in range(0,nStepsInTime):
    T[n,0] = 0
    T[n,N+1] = T[n,N-1] - beta*2*hx*(T[n,N]**4.0-T_inf**4.0)
    for i in range(1,N+1):
        T[n+1] = T[n,i] + ht*alpha/hx/hx*(T[n,i+1]-2*T[n,i]+T[n,i-1])
print(T)
```

```python
plt.plot(x,T[1])
```

```python
plt.imshow(T,interpolation = "bilinear")
plt.colorbar()
```
