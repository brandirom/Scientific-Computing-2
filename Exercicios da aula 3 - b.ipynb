```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

# Resolução da equação de van der pol, por Foward Euler:

```python
def van_der_pol(u_inicial, v_inicial, epsilon, passo=0.1, qnt_valores=101):
    def u_n(u, v):
        return v
    def v_n(u, v, epsilon):
        return (-epsilon*((u**2)-1)*v)-u
    val_u = [u_inicial]
    val_v = [v_inicial]
    for i in range(1, qnt_valores):
        val_v_1= val_v[i - 1] + passo * v_n(val_u[i - 1], val_v[i - 1], epsilon)
        val_u_1 = val_u[i - 1] + passo * u_n(val_u[i - 1], val_v_1)
        val_u.append(val_u_1)
        val_v.append(val_v_1)
    return val_u, val_v
```

# grafico do plano t x u:

```python
epsilon = [0.01,0.1,0.5,1.0,1.5,2.0,2.5,3.0,3.5,4.0]
for epsilons in epsilon:
    val = van_der_pol(1,0,epsilons)
    valores_u = val[0]
    plt.plot(valores_u,label=f"epsilon:{epsilons}")
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.title("T x U")
```

# grafico do plano u x v:

```python
epsilon = [0.01,0.1,0.5,1.0,1.5,2.0,2.5,3.0,3.5,4.0]
for epsilons in epsilon:
    van = van_der_pol(1,0,epsilons,0.1)
    valores_u = van[0]
    valores_v = van[1]
    plt.plot(valores_u,valores_v,label=f"epsilon:{epsilons}")
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.title("U x V")
```
