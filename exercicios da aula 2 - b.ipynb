```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

# du/dt=-5u, metodo Crank-Nicolson

```python
def exponenciais(n, p):
    lista = []
    pontos = np.linspace(0, n, p)    
    for ponto in pontos:
        lista.append(np.exp(-5 * ponto))        
    return lista
def aproximacao(valor_inicial, p, passo=0.1):
    valores_aproximados = [valor_inicial]
    for i in range(1, p):
        novo_valor = valores_aproximados[i-1] + (passo/2)*((-5*valores_aproximados[i-1])+(valores_aproximados[i-1]/6))
        valores_aproximados.append(novo_valor)
    return valores_aproximados 
```

```python
pontos = np.linspace(0, 5, 50)   
aproximacao = aproximacao(1,50,0.1)
exponenciais = exponenciais(5,50)
erro = [abs(abs(exponenciais[i])-abs(aproximacao[i])) for i in range(1,len(aproximacao))]
fig,(x1,x2)=plt.subplots(1,2)
x1.plot(pontos,exponenciais)
x1.plot(pontos,aproximacao)
x2.plot(pontos[1:],erro)
x1.set_title("grafico das funções")
x2.set_title("erro")
```

# du/dt=-5u, metodo Heun

```python
def exponenciais(n, p):
    lista = []
    pontos = np.linspace(0, n, p)    
    for ponto in pontos:
        lista.append(np.exp(-5 * ponto))        
    return lista
def equacao(t,u):
    return -5*u
def aproximacao(valor_inicial, p, passo=0.1):
    valores_aproximados = [valor_inicial]
    for i in range(1, p):
        t_atual = passo*i
        f_n=-5*valores_aproximados[i-1]
        novo_valor = valores_aproximados[i-1] + (passo/2)*(f_n+equacao(t_atual,valores_aproximados[i-1]+passo*f_n))
        valores_aproximados.append(novo_valor)
    return valores_aproximados 
```

```python
pontos = np.linspace(0, 5, 50)   
aproximacao = aproximacao(1,50,0.1)
exponenciais = exponenciais(5,50)
erro = [abs(abs(exponenciais[i])-abs(aproximacao[i])) for i in range(1,len(aproximacao))]
fig,(x1,x2)=plt.subplots(1,2)
x1.plot(pontos,exponenciais)
x1.plot(pontos,aproximacao)
x2.plot(pontos[1:],erro)
x1.set_title("grafico das funções")
x2.set_title("erro")
```
