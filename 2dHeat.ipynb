# 2dHeat

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def kappa1(x,y):
    return 410

def kappa2(x,y):
    if x<0.5:
        return 0.1
    else:
        return 410.0

def kappa3(x,y):
    if (x-0.05)**2 + (y-0.03)**2 < 0.3**2:
        return 410
    else:
        return 0.1
```

```python
def twodHeat(kappa):
    N = 99
    hx, hy = 1.0 / (N+1) , 1.0 / (N-1)
    T_left, T_right = 500.15,293.15
    A = np.zeros(((N*(N+2),N*(N+2))))
    f = np.zeros(N*(N+2))
    for j in range(1,N+1):
        for i in range(1, N+1):
            n = i + j * N -1
            xi = i*hx
            yj = (j-1) * hy
            kxp = kappa(xi+hx/2 , yj)
            kxm = kappa(xi-hx/2 , yj)
            kyp = kappa(xi, yj+hy/2)
            kym = kappa(xi, yj-hy/2)

            A[n,n] = -(kxp+kxm)/hx/hx - (kyp+kym)/hy/hy
            if i>1:
                A[n,n-1] = kxm/hx/hx
            else:
                f[n] = f[n] - T_left*kxm/hx/hx
            if i<N:
                A[n,n+1] = kxp/hx/hx 
            else:
                f[n] = f[n] - T_right*kxp/hx/hx
            A[n,n+N] = kyp/hy/hy
            A[n,n-N] = kym/hy/hy
    for i in range (1,N+1):
        n = i + 0 * N -1
        A[n,n] = 1.0
        A[n,n+2*N] = -1.0
    for i in range(1,N+1):
        n = i + (N+1) * N - 1
        A[n,n] = 1.0
        A[n,n-2*N] = -1.0
    T = np.linalg.solve(A,f)
    T = T.reshape((N+2,N))
    return T
```

```python
T = twodHeat(kappa1)
plt.imshow(T,interpolation = "bilinear")
plt.colorbar()
```

```python
T = twodHeat(kappa2)
plt.imshow(T,interpolation = "bilinear")
plt.colorbar()
```

```python
T = twodHeat(kappa3)
plt.imshow(T,interpolation = "bilinear")
plt.colorbar()
```
