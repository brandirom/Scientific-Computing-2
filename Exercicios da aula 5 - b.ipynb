# Estabilidade do MÃ©todo de Range-Kuta 4:

1. Let $\lambda \in \mathbb{C}$.

2. The derivative of $u$ with respect to time $t$ is given by $\frac{du}{dt} = \lambda u$.

3. The initial condition is $u(0) = 1$ for the sake of simplicity.  

4. Now we define:
 1. $$k_1 = \lambda u_n h $$    
 2. $$k_2 = \lambda(u_n + \frac{k_1}{2}) h$$ 
 3. $$k_3 = \lambda(u_n + \frac{k_2}{2}) h$$  
 4. $$k_4 = \lambda(u_n + k_3) h$$  

And:  
$$u_{n+1} = u_n + \frac{h}{6}(k_1 + 2k_2 + 2k_3 + k_4)$$  

Simplifying, we obtain:  
$$u_{n+1} = u_n.(1 + \lambda h + \frac{\lambda^2 h^2}{2!} + \frac{\lambda^3 h^3}{3!} + \frac{\lambda^4 h^4}{4!})$$

For the stability of the method, we have:
$$u_{n+1} = u_n e^{\lambda h} = R u_n$$  
TThen, for the stability of the method, we need $||R<1||$:  
Therefore:
$$||1 + \lambda h + \frac{\lambda^2 h^2}{2!} + \frac{\lambda^3 h^3}{3!} + \frac{\lambda^4 h^4}{4!})|| \leq 1$$

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-3, 3, 400)
y = np.linspace(-3, 3, 400)
X, Y = np.meshgrid(x, y)
Z = X + 1j*Y
expression = np.abs(1 + Z + Z**2/2 + Z**3/6 + Z**4/24)
plt.contour(X, Y, expression, levels=[1], colors='gray')

plt.xlim(-3, 3)
plt.ylim(-3, 3)
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```
