# Exercise 3.3.10:
Verify that the constant vector c in $\frac{dùõøu}{dt} = Jùõøu + c$ does not affect the stability analysis, i.e., the stability results are the same as for the homogeneous ODE system.  
Hint: Let $ùõøv = ùõøu + J^{‚àí1}c$ and formulate $\frac{dùõøu}{dt} = Jùõøu + c$ for $ùõøv$.

Let's start with the given equations:  
$ùõøv = ùõøu + J^{‚àí1}c \tag{1}$
and:  
$\frac{dùõøu}{dt} = Jùõøu + c \tag{2}$  
When we differentiate equation (1) with respect to time, we obtain:  
$$\frac{dùõøv}{dt} = \frac{dùõøu}{dt} + \frac{d(J^{‚àí1}c)}{dt}$$
Now, substituting equation (2) into the above expression, we get:  
$\frac{dùõøv}{dt} = Jùõøu + c + \frac{d(J^{‚àí1}c)}{dt} \tag{3}$
Next, let's isolate ùõøu from equation (1):  
$$ùõøu = ùõøv - J^{‚àí1}c$$
Substituting this into equation (3), we have:  
$$\frac{dùõøv}{dt} = J(ùõøv - J^{‚àí1}c) + c + \frac{d(J^{‚àí1}c)}{dt}$$
$$= Jùõøv - JJ^{‚àí1}c + c + \frac{d(J^{‚àí1}c)}{dt}$$
$$= Jùõøv - c + c + \frac{d(J^{‚àí1}c)}{dt}$$
$$= Jùõøv + \frac{d(J^{‚àí1}c)}{dt}$$
Since $J$ is a constant and does not depend on time, it can be taken out of the derivative:  
$$\frac{dùõøv}{dt} = Jùõøv + J^{‚àí1}\frac{dc}{dt}$$
Given that $c$ is independent of time, $\frac{dc}{dt} = 0$, simplifying the equation further:  
$$\frac{dùõøv}{dt} = Jùõøv$$

# Exercise 3.3.11:
Write a program to compute and plot the stability region SH for Heun‚Äôs method: $u_{k+1} = u_k + \frac{h}{2}(f(t_k, u_k) + f(t_{k+1}, u_k + hf(t_k, u_k)))$.

1. Let $\lambda \in \mathbb{C}$.

2. The derivative of $u$ with respect to time $t$ is given by $\frac{du}{dt} = \lambda u$.

3. The initial condition is $u(0) = 1$.

Using this information, we derive the following recursion for $u_{k+1}$:

$$u_{k+1} = u_k + \frac{h}{2}(\lambda u_k + \lambda(u_k + h\lambda u_k))$$  
Simplifying, we obtain:  
$$u_{k+1} = u_k(1 + h\lambda + \frac{(h\lambda)^2}{2})$$
4. Since $u(0) = 1$, we can express the general solution as:
$$u_n = (1 + h\lambda + \frac{(h\lambda)^2}{2})^n$$

For the stability of the method, we need:
$$\lim_{t \to \infty}u(t) = 0$$

As we have $\lambda \in \mathbb{C}$, we just need to plot:
$$||1 + h\lambda + \frac{(h\lambda)^2}{2}|| \leq 1$$

# exercise 3.3.13:
 Write a program that visualizes the numerical solution of the
trapezoidal method applied to the scalar model equation for some negative values
of $\lambda$

```python
def TCN(u0, l, t=0, p=200, h=0.005):
    u = [u0]
    for t in range(0,p+1):
        u_1 = (u0 + (u0*h*l)/2)/(1-(l*h)/2)
        u.append(u_1)
        u0 = u[-1]
    return u
```

```python
lambdas = [-9, -7, -6, -3, -1]
u0 = 1
for l in lambdas:
    u = TCN(u0, l)
    plt.plot(u, label=f'lambda = {l}')
    plt.legend()
plt.show()
```
