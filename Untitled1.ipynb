```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation
from IPython.display import HTML
```

```python
Nx = 100
Ny = 50
Lx = 2.0
Ly = 1.0
hx = Lx / (Nx+1)
hy = Ly / (Ny+1)
Tf = 2.6
ht = min(0.5*hx*hx,0.5*hy*hy)*0.9
Nt = int(Tf/ht)

T = np.zeros((Ny+2,Nx+2,Nt+1))

alpha = 15.0e-4

sx = alpha*ht / hx / hx
sy = alpha*ht / hy  / hy 

#initial Condition:
for j in range(0, Ny+2):
    for i in range(0, Nx+2):
        xi, yi = i * hx, j * hy
        T[j, i, 0] = 200.0
        if ((xi - 0.3)**2 + (yi - 0.5)**2 < 0.2**2) or (0.3 <= xi <= 0.7 and 0.3 <= yi <= 0.7) or (0.6 <= xi <= 0.7 and 0.7 <= yi <= 0.9):
            T[j, i, 0] = 400.0

        # Verifica se o ponto está dentro da letra 'U'
        elif ((xi - 1.0)**2 + (yi - 0.5)**2 < 0.2**2) or (0.8 <= xi <= 1.2 and 0.7 <= yi <= 0.9) or (1.2 <= xi <= 1.6 and 0.7 <= yi <= 0.9) or (0.8 <= xi <= 1.6 and 0.3 <= yi <= 0.4):
            T[j, i, 0] = 400.0

        # Verifica se o ponto está dentro da letra 'M'
        elif (0.9 <= xi <= 1.1 and 0.3 <= yi <= 0.7) or (1.3 <= xi <= 1.5 and 0.3 <= yi <= 0.7) or (0.9 <= xi <= 1.5 and 0.7 <= yi <= 0.9):
            T[j, i, 0] = 400.0

        # Verifica se o ponto está dentro da letra 'E'
        elif (0.8 <= xi <= 1.6 and 0.3 <= yi <= 0.4) or (0.8 <= xi <= 1.6 and 0.5 <= yi <= 0.6) or (0.8 <= xi <= 1.6 and 0.7 <= yi <= 0.8):
            T[j, i, 0] = 400.0

plt.imshow(T[:,:,0],interpolation = 'bilinear')
plt.colorbar()
```

```python
for n in range(Nt):
    if (n%1000 == 0):
        print("step = ",n)
    T[0,:,n+1] = 200.0
    T[Ny+1,:,n+1] = 200.0
    T[:,0,n+1] = 200.0
    T[:,Nx+1,n+1] = 200.0
    for j in range(1,Ny+1):
        for i in range(1,Nx+1):
            T[j,i,n+1] = sx*T[j,i+1,n] + sx*T[j,i-1,n] + sy*T[j+1,i,n] + sy*T[j-1,i,n] + (1-2*sx-2*sy)*T[j,i,n]

plt.imshow(T[:,:,0],interpolation = 'bilinear')
plt.colorbar()
```
