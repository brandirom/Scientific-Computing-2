```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

# metodo do ponto fixo + Backward Euler:

```python
def equacao(t,u):
    return (u-t)/(u+t)
def fix_point_and_backward(valor_inicial,p ,n=20 ,passo=0.1):
    valores_aproximados = [valor_inicial]
    t_atual = passo
    for i in range(1,p):
        phi = [valores_aproximados[i-1] + passo * equacao(passo, valores_aproximados[i-1])]
        for j in range(1,n):
            valor_phi = valores_aproximados[i-1] + passo * equacao(t_atual, phi[j-1])
            phi.append(valor_phi)
        valores_aproximados.append(phi[-1])
        t_atual += passo
    return valores_aproximados
teste_1=fix_point_backward(1, 65)
pontos = np.linspace(0, 6, 65)  
plt.plot(pontos,teste_1)
plt.title('grafico da solução numérica:')
plt.axhline(y=0, color='black', linewidth=0.8)
plt.axvline(x=0, color='black', linewidth=0.8)
```
