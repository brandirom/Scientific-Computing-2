```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

# Exemplo 1:

```python
def f(t,u):
    dy1 = 1.2*u[0] - 0.6 * u[0] * u[1]
    dy2 = 0.8 * u[0] + 0.3 * u[0] * u[1]
    return [dy1, dy2]

def range_kuta(u ,p ,h=0.1):
    t = p*h
    k1 = f(t,u)
    k2 = f(t+(h/2), [u[i] + (h/2)*k1[i] for i in range(len(u))])
    k3 = f(t+(h/2), [u[i] + (h/2)*k2[i] for i in range(len(u))])
    k4 = f(t+h, [u[i] + h*k3[i] for i in range(len(u))])
    u_1 = [u[i] + (h/6)*(k1[i] + 2*k2[i] + 2*k3[i] + k4[i]) for i in range(len(u))]
    return u_1
```

```python
ans = [[2, 1]]

for t in range(1, 20):
    u0 = ans[-1]
    u1 = range_kuta(u0, t)
    ans.append(u1)

u1 = [x[0] for x in ans]
u2 = [x[1] for x in ans]
t = np.arange(20)
fig,(x1,x2) = plt.subplots(1,2)
x1.plot(t, u1)
x1.set_xlabel('t')
x1.set_title("U 1:")
x2.plot(t, u2)
x2.set_title("U 2:")
x2.set_xlabel('t')
plt.show()
```
