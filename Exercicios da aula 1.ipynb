```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

# Quest√£o 1:

```python
def exponenciais(n, p):
    lista = []
    pontos = np.linspace(0, n, p)
    for ponto in pontos:
        lista.append(np.exp(ponto))
    return lista
def aproximacao(valor_inicial, p, passo=0.1):
    valores_aproximados = [valor_inicial]
    for i in range(1, p):
        novo_valor = valores_aproximados[i-1] + passo * valores_aproximados[i-1]
        valores_aproximados.append(novo_valor)
    return valores_aproximados
pontos = np.linspace(0, 1, 10)  
aproximacao = aproximacao(1,10)
exponenciais = exponenciais(1,10)
erro=[abs(exponenciais[i])-abs(aproximacao[i]) for i in range(1,len(aproximacao))]
fig,(x1,x2)=plt.subplots(1,2)
x1.plot(pontos,exponenciais)
x1.plot(pontos,aproximacao)
x2.plot(pontos[1:],erro)
x1.set_title("grafico das fun√ß√µes")
x2.set_title("erro")
```

# Quest√£o 2:

```python
def exponenciais_2(n, p):
    lista = []
    pontos = np.linspace(0, n, p)    
    for ponto in pontos:
        lista.append(np.exp(-5 * ponto))        
    return lista
def aproximacao_2(valor_inicial, p, passo=0.1):
    valores_aproximados = [valor_inicial]
    for i in range(1, p):
        novo_valor = valores_aproximados[i-1] - 5 * passo * valores_aproximados[i-1]
        valores_aproximados.append(novo_valor)
    return valores_aproximados 
pontos_2 = np.linspace(0, 1, 10)   
aproximacao_2 = aproximacao_2(1,10)
exponenciais_2 = exponenciais_2(1,10)
erro_2=[abs(exponenciais_2[i])-abs(aproximacao_2[i]) for i in range(1,len(aproximacao_2))]
fig,(x1,x2)=plt.subplots(1,2)
x1.plot(pontos_2,exponenciais_2)
x1.plot(pontos_2,aproximacao_2)
x2.plot(pontos_2[1:],erro_2)
x1.set_title("grafico das fun√ß√µes")
x2.set_title("erro")
```

```python
def equacao(t,u):
    return (u - t) / (u + t)
def aproximacao_3(valor_inicial, p ,n=0.01,passo=0.1):
    valores_aproximados = [valor_inicial]
    for i in range(1,p):
        phi = [valores_aproximados[j-1] + passo * equacao(t_atual, valores_aproximados[i-1])]
        t_atual = passo
        while abs(valores_aproximados[j-1]-phi)>n:
            phi = valores_aproximados[j-1] + passo * equacao(t_atual, phi[i-1])
            t_atual += passo
        valores_aproximados.append(phi[-1])
    return valores_aproximados
aproximacao_3 = aproximacao_3(1,70)
pontos=np.linspace(0, 7, 70)  
plt.plot(pontos,aproximacao_3)
plt.title('grafico da solu√ß√£o numerica:')
plt.axhline(y=0, color='black', linewidth=0.8)
plt.axvline(x=0, color='black', linewidth=0.8) 
```

```python
def exponenciais_4(n, p):
    lista = []
    pontos = np.linspace(0, n, p)    
    for ponto in pontos:
        lista.append(np.exp(-5 * ponto))        
    return lista
def aproximacao_4(valor_inicial, p, passo=0.1):
    valores_aproximados = [valor_inicial]
    for i in range(1, p):
        novo_valor = valores_aproximados[i-1] / (1+(5*passo)) 
        valores_aproximados.append(novo_valor)
    return valores_aproximados
```

```python
pontos_4_1 = np.linspace(0, 5.33, 13)   
aproximacao_4_1 = aproximacao_4(1,13,0.41)
exponenciais_4_1 = exponenciais_4(5.33,13)
erro_4_1=[abs(exponenciais_4_1[i])-abs(aproximacao_4_1[i]) for i in range(1,len(aproximacao_4_1))]
fig,(x1,x2)=plt.subplots(1,2)
x1.plot(pontos_4_1,exponenciais_4_1)
x1.plot(pontos_4_1,aproximacao_4_1)
x2.plot(pontos_4_1[1:],erro_4_1)
x1.set_title("grafico das fun√ß√µes com h=0.41")
x2.set_title("erro")
```
